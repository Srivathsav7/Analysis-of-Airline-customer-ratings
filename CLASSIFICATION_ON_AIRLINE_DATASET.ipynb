```python
#Only considering variables with correlation > 0.3
#Because the maximum correlation that exists is 0.5, picking the range of 0.3-0.5 for getting the predictions and fitting the model
df1= df[['Cleanliness','Leg room service','On-board service','Inflight entertainment','Seat comfort','Type of Travel','Class','Online boarding','satisfaction']]
```

```python
X= df1[['Cleanliness','Leg room service','On-board service','Inflight entertainment','Seat comfort','Type of Travel','Class','Online boarding']]
y=df1[['satisfaction']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#Was run for k until the nearest neighbours are 20, more iterations can be done 
from sklearn.neighbors import NearestNeighbors, KNeighborsClassifier
from sklearn.metrics import accuracy_score
# Train a classifier for different values of k
results = []
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train.values.ravel())
    results.append({
    'k': k,
    'accuracy': accuracy_score(y_test, knn.predict(X_test))
 })

# Convert results to a pandas data frame
results = pd.DataFrame(results)
print(results)
```

```python
plt.plot(results['k'],results['accuracy'])
plt.title('Determining the ideal k')
plt.xlabel('k-value')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.metrics import plot_confusion_matrix
knn = KNeighborsClassifier(n_neighbors=17).fit(X_train, y_train)
y_pred1=knn.predict(X_test)
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred1)
print(confusion_matrix)
a=confusion_matrix
plot_confusion_matrix(knn, X_test, y_test) 
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
Sensitivity2=a[0,0]/(a[0,0]+a[0,1])
print("Sensitivity:",Sensitivity2)
Specificity2=a[1,1]/(a[1,0]+a[1,1])
print("Specificity:",Specificity2)
metrics.plot_roc_curve(knn, X_test, y_test)
plt.show()
```

```python
#for applying logistic regression, applying a different method to determine the features and also created dummy variables
#reloading the dataset again
df1 = pd.read_csv('C:\\Users\\Srivathsav\\Full.csv') 
```

```python
arrivalFilter = lambda df1: df1[np.isnan(df1['Arrival Delay in Minutes']) == False]
df1 = arrivalFilter(df1)
df1.isnull().sum()
```

```python
#Dummy variables approach for logistic regression
cat_vars=['Gender','Customer Type','Type of Travel','Class']
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(df1[var], prefix=var)
    df2=df1.join(cat_list,how='left', lsuffix='_left', rsuffix='_right')
    df1=df2
cat_vars=['Gender','Customer Type','Type of Travel','Class']
df1_vars=df1.columns.values.tolist()
to_keep=[i for i in df1_vars if i not in cat_vars]
data_final=df1[to_keep]
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data_final['satisfaction']= le.fit_transform(data_final.satisfaction.values)
```

```python
#Feature selection for the model
#Done based on the chi-square score
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
y1=data_final[['satisfaction']]
X1=data_final.drop(['satisfaction'],axis=1)
fit = bestfeatures.fit(X1,y1)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X1.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```
